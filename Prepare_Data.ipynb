```python
import pandas as pd
```

```python
main_data = pd.read_csv('shootings.csv')
main_data
```

Process for stack area chart

```python
df1 = pd.DataFrame(columns = ['year', 'total_case', 'unarmed_case'])
years = [2015,2016,2017,2018,2019]
for i in range(5):
    df1.loc[i] = {'year': years[i], 'total_case': len(main_data[main_data.year==years[i]]),\
                  'unarmed_case': len(main_data[(main_data.year==years[i])&(main_data.armed=='unarmed')])}
df1
    
```

```python
df1.to_csv('stack_area_chart_data.csv')
```

Process for map and bar chart

```python
df2 = pd.DataFrame(columns = ['key', 'year', 'state', 'total_case', 'unarmed_case'])
years = [2015,2016,2017,2018,2019]
states = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
           'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
           'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
           'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
           'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
for year in years:
    for state in states:
       df2 = df2.append({'key': str(year)+state, 'year': year, 'state': state, \
                    'total_case': len(main_data[(main_data.year==year)&(main_data.state==state)]), \
                   'unarmed_case': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.armed=='unarmed')]) },\
                  ignore_index='True')
df2
```

```python
df2.to_csv('map_and_bar_data.csv')
```

Recorded vs. Unrecorded, state wise and national

```python
#State wise
df3 = pd.DataFrame(columns = ['key', 'year', 'state', 'recorded', 'unrecorded'])
years = [2015,2016,2017,2018,2019]
states = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
           'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
           'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
           'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
           'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
for year in years:
    for state in states:
        df3 = df3.append({'key': str(year)+state, 'year': year, 'state': state, \
                    'recorded': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.body_camera==True)]), \
                   'unrecorded': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.body_camera==False)]) },\
                  ignore_index='True')
df3
```

```python
df3.to_csv('if_recorded_state_wise.csv')
```

```python
#National
df4 = pd.DataFrame(columns = ['year','recorded', 'unrecorded'])
years = [2015,2016,2017,2018,2019]

for year in years:
        df4 = df4.append({'year': year,\
                    'recorded': len(main_data[(main_data.year==year)&(main_data.body_camera==True)]), \
                   'unrecorded': len(main_data[(main_data.year==year)&(main_data.body_camera==False)]) },\
                  ignore_index='True')
df4
```

```python
df4.to_csv('if_recorded_national.csv')
```

Prepare for ethnicity, state wise vs. national

```python
df5 = pd.DataFrame(columns = ['key', 'year', 'state', 'Asian', 'Black', 'Hispanic', 'Native', 'Other', 'White'])
years = [2015,2016,2017,2018,2019]
states = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
           'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
           'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
           'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
           'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
for year in years:
    for state in states:
        df5 = df5.append({'key': str(year)+state, 'year': year, 'state': state, \
                    'Asian': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.race=='Asian')]), \
                        'Black': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.race=='Black')]) ,\
                         'Hispanic': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.race=='Hispanic')]) ,\
                         'Native': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.race=='Native')]) ,\
                         'Other': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.race=='Other')]) ,\
                         'White': len(main_data[(main_data.year==year)&(main_data.state==state)&(main_data.race=='White')]) },\
                  ignore_index='True')
df5
```

```python
df5.to_csv('ethnicity_state_wise.csv')
```

```python
#National
df6 = pd.DataFrame(columns = ['year','Asian', 'Black', 'Hispanic', 'Native', 'Other', 'White'])
years = [2015,2016,2017,2018,2019]

for year in years:
        df6 = df6.append({'year': year, \
                    'Asian': len(main_data[(main_data.year==year)&(main_data.race=='Asian')]), \
                        'Black': len(main_data[(main_data.year==year)&(main_data.race=='Black')]) ,\
                         'Hispanic': len(main_data[(main_data.year==year)&(main_data.race=='Hispanic')]) ,\
                         'Native': len(main_data[(main_data.year==year)&(main_data.race=='Native')]) ,\
                         'Other': len(main_data[(main_data.year==year)&(main_data.race=='Other')]) ,\
                         'White': len(main_data[(main_data.year==year)&(main_data.race=='White')]) },\
                  ignore_index='True')
df6
```

```python
df6.to_csv('ethnicity_national.csv')
```
